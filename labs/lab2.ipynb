## Problem 1: The Doubling Machine

A list comprehension is a one-line shortcut for building a list.

Below, the `for` loop builds a list of every number **doubled**. Your job: do the same thing in **one line** using a list comprehension.

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# The long way (already done for you):
doubled_loop = []
for num in numbers:
    doubled_loop.append(num * 2)
print("With a loop:", doubled_loop)

# The short way — one line!
doubled = [x*2 for x in numbers]

print("With a comprehension:", doubled)
```

```python
# ✅ CHECK CELL — just run me!
assert doubled == [6, 2, 8, 2, 10, 18, 4, 12], "Not quite — did you multiply each number by 2?"
print("✅ Problem 1 complete! The Doubling Machine works!")
```

## Problem 2: Picky Doubling
Now make the machine picky: double **only the odd numbers**, and skip the rest.

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

odd_doubled = [x*2 for x in numbers if x%2==1]

print(odd_doubled)
```

```python
# ✅ CHECK CELL — just run me!
assert odd_doubled == [6, 2, 2, 10, 18], "Only the odd numbers should be doubled — the evens get skipped entirely!"
print("✅ Problem 2 complete! Picky machine calibrated.")
```

## Problem 3: Secret Message Decoder
A spy sent you a message, but they hid it by adding a fake letter **after every real letter**.

So `"hzezlzlzo"` is really `"hello"` — you just need every **other** character, starting from position 0.

```python
def decode(message):
  decoded=""
  for i in range(0,len(message),2):
    decoded+=message[i]
  return decoded

secret = "pzyztzhzoznzizszazwzezszozmze"
print(decode(secret))  # what does it say?
```

```python
assert decode("hzezlzlzo") == "hello", "Try slicing with a step of 2!"
assert decode("czazmzp") == "camp"
print("✅ Problem 3 complete! You cracked the code, agent. 🕵️")
```

## Problem 4: Palindrome Detector (≈8 min)

A **palindrome** reads the same forwards and backwards: `racecar`, `level`, `noon`.

Write a function that returns `True` if a word is a palindrome, `False` if not.

⚠️ Make it ignore capitalization — `"Racecar"` should still count!

```python
def is_palindrome(word):
    backwards=word[::-1]
    return word.lower()==backwards.lower()

print(is_palindrome("racecar"))  # should be True
print(is_palindrome("python"))   # should be False
```

```python
assert is_palindrome("noon") == True
assert is_palindrome("Racecar") == True, "Don't forget .lower()!"
assert is_palindrome("data") == False
print("✅ Problem 4 complete! taco cat approves. 🌮🐱")
```

## Problem 5: The Vowel Counter

Write a function that counts how many **vowels** (a, e, i, o, u) are in a sentence.

⚠️ Capital vowels count too!

```python
def count_vowels(sentence):
    sentence2=sentence.lower()
    count=0
    for char in sentence2:
        if char in "aeiou":
            count+=1
    return count

print(count_vowels("I love data science"))  # should be 8
```

```python
assert count_vowels("hello") == 2
assert count_vowels("AEIOU") == 5, "Capital vowels count too — try .lower()!"
assert count_vowels("rhythm") == 0
print("✅ Problem 5 complete! You can count on yourself. 🔢")
```

## Problem 6: Mini Data Analysis — Heat Wave

Your first taste of real data science! Here are daily high temperatures (°F) for two weeks in Ann Arbor.

Compute three things **without** typing any numbers by hand:
1. `hottest` — the highest temperature
2. `average` — the mean temperature (sum ÷ how many)
3. `hot_days` — how many days were **above 85°F**

💡 *Hint: Python gives you* `max()`, `sum()`, *and* `len()` *for free.*

💡 *Hint for #3: build a filtered list with a comprehension, then take its* `len()` *— just like Problem 2!*

```python
temps = [78, 85, 91, 88, 76, 82, 95, 89, 84, 79, 92, 87, 81, 90]

hottest = max(temps)
average = sum(temps)/len(temps)
hot_days = len([x for x in temps if x>85])

print("Hottest day:", hottest)
print("Average temp:", average)
print("Days above 85:", hot_days)
```

```python
assert hottest == 95
assert abs(average - 85.5) < 0.01, "Average = sum of temps ÷ number of temps"
assert hot_days == 7, "Count days STRICTLY above 85 — 85 itself doesn't count!"
print("✅ Problem 6 complete! That's data analysis — for real! 📊")
```

## Problem 7: FizzBuzz, Camp Edition

A classic! Write a function that takes a number `n` and returns:
- `"Fizz"` if it's divisible by 3
- `"Buzz"` if it's divisible by 5
- `"FizzBuzz"` if it's divisible by **both**
- otherwise, the number itself (as a number, not a string)

⚠️ Watch the order of your checks — what happens to 15 if you check "divisible by 3" first?

💡 *Hint:* `n % 3 == 0` *means "n is divisible by 3".*

```python
def fizzbuzz(n):
    if n%3==0 and n%5==0:
      return "FizzBuzz"
    elif n%3==0:
      return "Fizz"
    elif n%5==0:
      return "Buzz"
    else:
      return n

# Watch it run on 1 through 15:
for i in range(1, 16):
    print(i, "→", fizzbuzz(i))
```

```python
assert fizzbuzz(9) == "Fizz"
assert fizzbuzz(10) == "Buzz"
assert fizzbuzz(15) == "FizzBuzz", "Check for 'both' FIRST, or 15 gets caught by the 3-check!"
assert fizzbuzz(7) == 7
print("✅ Problem 7 complete! You passed a real interview question. 😎")
```

## 🌟 Bonus 1: Password Strength Checker

Write a function that rates a password:
- `"weak"` if it's shorter than 8 characters
- `"okay"` if it's 8+ characters
- `"strong"` if it's 8+ characters **and** contains at least one digit

💡 *Hint:* `any(ch.isdigit() for ch in password)` *tells you if there's a digit anywhere.*

```python
def password_strength(password):
    if len(password)>8 and any(ch.isdigit() for ch in password):
      return "strong"
    elif len(password)>8:
      return "okay"
    else: return "weak"

print(password_strength("cat"))          # weak
print(password_strength("ilovepython"))  # okay
print(password_strength("python4ever!")) # strong
```

```python
assert password_strength("abc") == "weak"
assert password_strength("longpassword") == "okay"
assert password_strength("longpassw0rd") == "strong"
print("🌟 Bonus 1 complete! You are officially a Python security expert.")
```
